This is the image algorithm



We use this to get user images to train our recognition model and prepare random images for training data

```python
#Training data for random people


import os
import cv2 as cv
import numpy as np

# Input directory containing images
input_dir = 'Temp'

# Output directory to save extracted faces
output_dir = 'Data\\Random_Faces'

# Initialize face detector (example using Haar cascades)
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Face ID counter
face_id = 0

# Iterate through each file in the input directory
for filename in os.listdir(input_dir):
    if face_id >= 1120:
                break
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        # Read the image
        image_path = os.path.join(input_dir, filename)
        frame = cv.imread(image_path)
        
        # Convert to grayscale for face detection
        gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        
        
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
        
        # Iterate over detected faces
        # Check if any faces are detected
        if len(faces) > 0:
            
            for (x, y, w, h) in faces:
                # Extract the detected face
                detected_face = frame[y:y+h, x:x+w]
                #0.2 is the lowest contrast 
                #0.07 is the lowest brightness
                
                
                if(face_id>=750):
                    detected_face=cv.GaussianBlur(detected_face, (31,31), 0)
                    brightness = 0.07
                    contrast = 0.2

                elif((face_id>=400)and(face_id<750)):
                     detected_face=cv.GaussianBlur(detected_face, (25,25), 0)
                     brightness = 0.63
                     contrast = 0.9
                elif(face_id<400):
                     detected_face=cv.GaussianBlur(detected_face, (17,17), 0)
                     brightness = 0.28
                     contrast = 0.5
            
                      
                      
                detected_face = cv.addWeighted(detected_face, contrast, np.zeros(detected_face.shape, detected_face.dtype), 0, brightness) 
                
                
                # Save the detected face
                face_filename = os.path.join(output_dir, f"face_{face_id}.jpg")
                cv.imwrite(face_filename, detected_face)
                
                # Increment face ID counter
                face_id += 1
                
                
```

```python


import os
import cv2 as cv

video=cv.VideoCapture(0)

face_id = 0

face_classifier = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

output_dir = 'Data//Detected_Faces'
os.makedirs(output_dir, exist_ok=True)


while(True):
    if face_id >= 1125:
                break
    ret,frame=video.read()


    grey=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(
    grey, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cv.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 4)

    for (x, y, w, h) in faces:
        
        detected_face = frame[y:y+h, x:x+w]

       
        face_filename = os.path.join(output_dir, f"face_{face_id}.jpg")
        cv.imwrite(face_filename, detected_face)

        # Increment face ID counter
        face_id += 1
    cv.imshow('frame',frame)
    if cv.waitKey(1) & 0xFF == ord('q'): 
        break

video.release() 

cv.destroyAllWindows() 
```
